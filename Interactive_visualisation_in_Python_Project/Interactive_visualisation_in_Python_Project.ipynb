###  Exercise 1

Create an interactive function called `show_articles_more_than`, that filters a dataset based on a selected column and a threshold value.

It should allow the user to:
* input a column name using a textbox
* choose a numeric threshold using a slider
* It should display a formatted heading showing the filter condition
* It should return a table of articles where the selected column value is greater than the chosen threshold
* The output should include the floowing columns title, claps, published_date, read_time, tags, views, and reads.

```python
@interact
def show_articles_more_than(column='claps', x=5000):
    display(HTML(f'<h3> Showing articles with more than {x} {column} <h3>'))
    display(df.loc[df[column] > x, ['title', 'claps', 'published_date', 'read_time', 'tags', 'views', 'reads']])
```

### Exercise 3

Create an interactive function called `correlations` that computes and displays the correlation between two columns in a dataset.

It should allow the user to:
* select a first numeric column from a dropdown.
* select a second numeric column from a dropdown.
* It should print the correlation value between the two selected columns in a formatted string.

```python
@interact
def correlations(column_1= list(df.select_dtypes('number').columns),
                 column_2= list(df.select_dtypes('number').columns)
                 ):
    
    if column_1 == column_2:
        print(f'Both columns should be different')
    else:
        print(f"Correlation value: {df[column_1].corr(df[column_2])}")
```

### Exercise 4

Create an interactive function called `scatter_plot` that generates a scatter plot between two numeric columns in a dataset using plotly.

It should allow the user to:
* select an x-axis column from a dropdown of numeric columns in the dataframe.
* select a y-axis column from a dropdown numeric columns, but defaulting to the second column in the list
* Generate a scatter plot

```python
@interact
def scatter_plot(x= list(df.select_dtypes('number').columns),
                y= list(df.select_dtypes('number').columns)[1:]):
    
    if x == y:
        print(f"Both columns should not be the same.")
    else:
        figure = px.scatter(df, x=x, y=y, title=f"{y.title()} vs {x.title()}")
        figure.show()
```

### Exercise 5

Create an interactive function called `plot_corrs` that generates a customisable annotated correlation heatmap using plotly.figure_factory.

It should:
* Define a list of colorscales called `cscales` containing the following options: `'Greys'`, `'YlGnBu'`, `'Greens'`, `'YlOrRd'`, `'Bluered'`, `'RdBu'`, `'Reds'`, `'Blues'`, `'Picnic'`, '`Rainbow'`, `'Portland'`, `'Jet'`, `'Hot'`, `'Blackbody'`, `'Earth'`, `'Electric'`, `'Viridis'`, `'Cividis'`
* Allow the user to select a colorscale from a dropdown populated with cscales
* Generate an annotated heatmap with correlation values rounded to 2 decimal places

```python
cscales = ['Greys', 'YlGnBu', 'Greens', 'YlOrRd', 'Bluered', 'RdBu', 
           'Reds', 'Blues', 'Picnic', 'Rainbow', 'Portland', 'Jet', 
           'Hot', 'Blackbody', 'Earth', 'Electric', 'Viridis', 'Cividis']

import plotly.figure_factory as ff

corrs = df.corr(numeric_only=True)

@interact
def plot_heatmap(colourscale=cscales):
    figure = ff.create_annotated_heatmap(z= corrs.round(2).values,
                                         x = list(corrs.columns),
                                         y = list(corrs.index),
                                         colorscale=colourscale,
                                         annotation_text= corrs.round(2).values)
    
    figure.show()
```
